```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Prende i dati da file e crea l'ascissa
data=np.loadtxt("BS.dat")
dim=np.shape(data)[0]
x=np.linspace(1,dim,dim)

#Plotto i dati simulati e teorici
plt.errorbar(x,data[:,0],data[:,4],label="Simulazione")
plt.plot(x,np.ones(dim)*14.975790778311286,label="Black-Scholes")
#Parte estetica
plt.title(f"Call-Option\nGeometric Brownian Motion")
plt.xlabel("Numero blocchi")
plt.ylabel(r"$\langle C[S(0),0]\rangle$")
plt.grid()
plt.legend()

plt.show()
```

```python
#Plotto i dati simulati e teorici
plt.errorbar(x,data[:,1],data[:,5],label="Simulazione")
plt.plot(x,np.ones(dim)*14.975790778311286,label="Black-Scholes")
#Parte estetica
plt.title(f"Call-Option\nDiscrete")
plt.xlabel("Numero blocchi")
plt.ylabel(r"$\langle C[S(0),0]\rangle$")
plt.grid()
plt.legend()

plt.show()
```
