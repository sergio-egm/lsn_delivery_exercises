```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=[]
suff=["epot","ekin","temp","etot","pres"]
tit=[
    "Energia Potenziale",
    "Energia Cinetica",
    "Temperatura",
    "Energia Totale",
    "Pressione"
]
for s in suff:
    data.append(np.loadtxt("../MD_MC/MDNVE_MCNVT/output_"+s+".dat"))
```

```python
fig,axx = plt.subplots(3,2,figsize=(12,10))
axx=np.reshape(axx,6)
x,y=(0,0)
i=0

for ax in axx[0:-1]:
    y=i%2
    x=int(i/2)
    ax.errorbar(data[i][:,0],data[i][:,2], yerr=data[i][:,3])
    ax.grid()
    ax.set(xlabel="Numero blocchi")
    ax.set_title(tit[i])
    i+=1
    
axx[-1].axis('off')
plt.show()
```
